```python
#Let's remove unnecessary columns
for column in df.columns:
    if df[column].nunique() == 1:
        del df[column]
```

```python
df
```

```python
#Let's normalize columns
#Let's check if the column has more than 2 unique values (non-binary)

for column in df.columns:
    if  df[column].nunique() != 2:
        df[column] = (df[column]-df[column].min())/(df[column].max() - df[column].min())
```

```python
df
```

```python
#Let's examine auto correlation
columns = set()
for column1 in df.columns:
    for column2 in df.columns:
        if column1 > column2 and df[column1] .corr(df[column2]) > 0.55:
            columns.add(column1)
            print(column1 , column2)
```

```python
#Let's remove auto correlation columns 
for column in list(columns):
    del df[column]
```

```python
#Let's import library
from sklearn.cluster import DBSCAN
```

```python
#Let's define X 
X = df
```

```python
#Let's set eps and min sample and fit model 
dbscan = DBSCAN(eps=1.3, min_samples=2)
labels = dbscan.fit_predict(X)
```

```python
print("Cluster labels:", labels)
print("Number of clusters:", len(np.unique(labels)) - 1)
```

```python
labels_series = pd.Series(labels)
count_minus_one = (labels_series == -1).sum()
```

```python
count_minus_one
```

```python
#Let's create as dynamic parametres
anomaly = []
epsilon = []
minsample = []
anomall = []
for eps in np.arange(1,2.1,0.2):
    for ms in range(1,10): 
        epsilon.append(eps)
        minsample.append(ms) 
        dbscan = DBSCAN(eps=eps, min_samples=ms)
        labels = dbscan.fit_predict(X)
        labels_series = pd.Series(labels)
        count_minus_one = (labels_series == -1).sum()
        anomall.append(count_minus_one) 
        if count_minus_one < len(df)/70 and count_minus_one > 0:
             print(f'Epsilon: {eps} \nMinimum Samples: {ms}')
             print('-----------------------------------------')
             print("Number of clusters:", len(np.unique(labels)) - 1)
             print(f'Anomaly count: {count_minus_one}')
             print('==========================================\n\n')
             anomaly.append(count_minus_one) 
```

```python
#Let's find average number of anomaly point
print(f"Average number of anomalies: {np.round(np.mean(anomaly),1)}")
```

<font size = 4>💡 Silhouette

```python
#Let's use silhouette method for finding the best parameters
from sklearn.metrics import silhouette_score

best_eps = 0
best_min_samples = 0
best_silhouette_score = -1

for eps in np.arange(1,2.1,0.2):
    for ms in range(1,10): 
        epsilon.append(eps)
        minsample.append(ms) 
        dbscan = DBSCAN(eps=eps, min_samples=ms)
        labels = dbscan.fit_predict(X)
        # Check if there are any clusters other than noise (-1)
        if len(set(labels)) > 1:
            silhouette_avg = silhouette_score(X, labels)
            if silhouette_avg > best_silhouette_score:
                best_silhouette_score = silhouette_avg
                best_eps = eps
                best_min_samples = ms

print(f"Best EPS: {best_eps}")
print(f"Best Min Samples: {best_min_samples}")
print(f"Best Silhouette Score: {best_silhouette_score}")
```
